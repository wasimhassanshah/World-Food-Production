```python
# Get the value counts for the 'Area' column and sort them in descending order
area_counts = dataset['Area'].value_counts().sort_values(ascending=False)

# Create a DataFrame from the value counts
df = pd.DataFrame({'Country': area_counts.index, 'Count': area_counts.values})

# Add a 'Rank' column starting from 1 to the number of rows
df['Rank'] = range(1, len(df) + 1)

# Reorder the columns
df = df[['Rank', 'Country', 'Count']]

# Display the top 20 rows
df
```

**Food Production of Countries Year Wise**

```python
#Set the size of the plot

 

# setting the size of the plot
plt.rcParams['figure.figsize'] = (20, 10)


# looking at India's Growth
india_production = pd.DataFrame(dataset[dataset['Area'] == 'India'].loc[:, "Y2003": "Y2013"].agg("sum", axis = 0))

india_production.columns = ['Production']
plt.subplot(231)
sns.barplot(x = india_production.index, y = india_production.Production, data = india_production)
plt.gca().set_title("India's Growth")



# looking at Pakistan's Growth
pak_production = pd.DataFrame(dataset[dataset['Area'] == 'Pakistan'].loc[:, "Y2003": "Y2013"].agg("sum", axis = 0))

pak_production.columns = ['Production']
plt.subplot(232)
sns.barplot(x = pak_production.index, y = pak_production.Production, data = pak_production)
plt.gca().set_title("Pakistan's Growth")


# looking at USA Growth
usa_production = pd.DataFrame(dataset[dataset['Area'] == 'United States of America'].loc[:, "Y2003": "Y2013"].agg("sum", axis = 0))

usa_production.columns = ['Production']
plt.subplot(233)
sns.barplot(x = usa_production.index, y = usa_production.Production, data = usa_production)
plt.gca().set_title("USA Growth")
```

**Plotting Anual Production of Crops for each COuntry**



```python
# plotting for the Annual Production of crops by every country

countries = list(dataset['Area'].unique())
years = list(dataset.iloc[:, 10:].columns)

plt.figure(figsize = (25, 10))
for i in countries:
  production = []
  for j in years:
    production.append(dataset[j][dataset['Area'] == i].sum())
  plt.plot(production, label = i)
  
plt.xticks(np.arange(53), tuple(years), rotation = 30)
plt.title('Country wise Annual Production')
plt.legend(bbox_to_anchor = (0., 1.02, 1.,  .102), loc = 3, ncol = 10, mode = 'expand', borderaxespad = 0.)
plt.savefig('p.png')
plt.show()
```

**Creating New data sets Containing Information about Countries and Production Only**

```python
new_data_dict = {}

for i in countries:
    production = []
    for j in years:
        production.append(dataset[j][dataset['Area']==i].sum())
    new_data_dict[i] = production
new_data= pd.DataFrame(new_data_dict)

new_data.head()
```

```python
# Making transpose of the new data and assigning year names to them

new_data = pd.DataFrame.transpose(new_data) 
new_data.columns = years


new_data.head()
```

```python
new_data.shape
```

```python
# Feature ENgineer Column

# Defining new column to view mean production country wise

mean_production = []

for i in range(174):
    mean_production.append(new_data.iloc[i,:].values.mean())
new_data['Mean Production'] = mean_production

# Defining rank according to mean production

new_data['Rank'] = new_data['Mean Production'].rank(ascending=False)

```
